```python
# checking the presence of outliers
def outliers_iqr(series):
  q1 = series.quantile(0.25)
  q3 = series.quantile(0.75)
  IQR  = q3-q1
  lower_range = q1-1.5*IQR
  higher_range = q3+1.5*IQR
  return(series < lower_range) | (series > higher_range)
outliers = green_dataset.select_dtypes(include=['number']).apply(outliers_iqr)
if outliers.any().any():
  print('outliers are present')
else:
  print('outliers are absent')
```

```python
# checking the presence of outliers in each column
def detect_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (series < lower_bound) | (series > upper_bound)

# Check for outliers in numeric columns
outliers = green_dataset.select_dtypes(include=['number']).apply(detect_outliers_iqr)

# Check if any outliers are found in each column
for column in outliers.columns:
    if outliers[column].any():
        print(f"Outliers are present in column '{column}'.")
```

```python
# visualizing the graph for box plot
# importing the libraries
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
green_dataset.boxplot(column='MonthlyIncome')
plt.title('Analyzing Monthly Income')
plt.ylabel('Values')
plt.xlabel('Money')
plt.show()
```

```python
def detect_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (series >= lower_bound) & (series <= upper_bound)

# Apply outlier detection and filter dataset to remove outliers
cleaned_dataset = green_dataset[green_dataset.select_dtypes(include=['number']).apply(detect_outliers_iqr).all(axis=1)]
cleaned_dataset.to_csv('cleanedgreen',index=False)
```

```python
# calculating the total attrition values
Attrition_total = cleaned1_data.Attrition.count()
Attrition_total
```

```python
# calculating only the true attrition values
Attrition_true = cleaned1_data[cleaned1_data['Attrition']=='Yes']
Attrition_true
```

```python
# visualization of attrition using pie chart

attrition_percentage = [total_attrition_rate, 100 - total_attrition_rate]
labels = ['peoples left', 'peoples in the company']

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(attrition_percentage, labels=labels, autopct='%1.1f%%', startangle=90, colors=['tab:blue', 'tab:orange'])
plt.title('Percentage of Attrition')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```
